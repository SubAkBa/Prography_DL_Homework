```python
from mtcnn.mtcnn import MTCNN
import cv2
import matplotlib.pyplot as plt

detector = MTCNN()

image = cv2.imread("avengers/evans.jpg")
result = detector.detect_faces(image)


bounding_box = result[0]['box']
# bounding_box[0] # 350
# bounding_box[1] # 114
# bounding_box[2] # 198
# bounding_box[3] # 260
# nrof_faces = bounding_box.shape[0]
# print(nrof_faces)

face_img = image[bounding_box[1] : bounding_box[1] + bounding_box[3], 
                 bounding_box[0] : bounding_box[0] + bounding_box[2]]

image[bounding_box[1] : bounding_box[1] + bounding_box[3], 
      bounding_box[0] : bounding_box[0] + bounding_box[2]] = cv2.blur(face_img, (25, 25))

# cv2.imwrite("evans_result.jpg", image)

plt.imshow(image)
plt.show()
```
